## Exploring Seaborn's Built-in Dataset

#### 0.1 EXPLORE CATEGORICAL DATA USING SEABORN

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Importons l'un des ensembles de données intégrés de Seaborn appelé Tips.
Il contient des données sur les personnes qui donnent des pourboires aux restaurants le jour.

```python
tips_dataset = sns.load_dataset('tips')
```

```python
tips_dataset.head()
```

Dans l'ensemble de données ci-dessus, notez les chiffres et les colonnes catégorielles.
Essayons maintenant un diagramme à barres où x est la donnée catégorielle et y la donnée numérique par défaut le total_bill avec représenté par une moyenne. Mais nous pouvons utiliser l'estimateur pour changer cela. Voir le prochain
ligne de codes.

```python
sns.barplot(x='sex', y='total_bill',data= tips_dataset)
```

Au lieu de la facture totale moyenne dans le graphique ci-dessus, indiquons l'écart type en utilisant
l'argument de l'"estimateur". C'est la raison pour laquelle nous avons importé Numpy au début.

```python
sns.barplot(x='sex', y='total_bill',data= tips_dataset, estimator= np.std)

#Vous pouvez voir ici que les variations du total_bill sont plus importantes pour les hommes que pour les femmes.
```

Maintenant, essayons un CountPlot(). Dans ce cas, vous n'avez qu'à transmettre un argument catégorique.

```python
sns.countplot(x ='sex',data = tips_dataset)

# Nous indiquons ici le nombre total pour chaque catégorie de sexe, homme et femme.
```

Passons maintenant à un exemple de BoxPlot() pour explorer la distribution des factures totales par jour.

```python
sns.boxplot(x = "day", y = 'total_bill' , data = tips_dataset)

# La facture totale médiane est proche de 20 dollars par jour à partir de l'ensemble des données.
# Les points en haut sont des valeurs aberrantes.
# La plus grande variation est observée le samedi, y compris la facture totale la plus basse de tous les jours.
```

Séparons chaque jour le client fumeur et le client non fumeur en utilisant l'argument "hue".

```python
sns.boxplot(x = "day", y = 'total_bill' , data = tips_dataset, hue = 'smoker')

#Ici, les non-fumeurs paient en moyenne plus que les fumeurs le vendredi.
# La variation (Max-Min) est plus élevée pour les fumeurs le vendredi.
```

Essayons un ensemble de données plus compliqué : le Violin Plot.

```python
sns.violinplot(x = 'day', y = 'total_bill' , data = tips_dataset)

# Les tracés au violon donnent une vision beaucoup plus claire de la densité des données.
# Vous pouvez voir où la plupart des gens paient environ 20 $ pour leur repas.
```

Séparons maintenant chaque intrigue au violon par la catégorie Homme/Femme.

```python
sns.violinplot(x = 'day', y = 'total_bill' , data = tips_dataset, hue = 'sex')

# Chaque catégorie a sa propre intrigue de violon. Si vous souhaitez les combiner, consultez le site next␣
#→line de code.
```

```python
sns.violinplot(x = 'day', y = 'total_bill' , data = tips_dataset, hue = 'sex', split = True)

# Maintenant, nous combinons Homme contre Femme dans une intrigue de Violon pour chaque jour.
# Cela vous donne une meilleure vue de la distribution des factures totales pour les hommes et les femmes.
```

Explorons le Strip Plot.

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips_dataset, jitter = False)

# Lorsque la jitter est réglée sur Faux, il est difficile d'examiner la densité des données.
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips_dataset, jitter = True)

# Maintenant, c'est beaucoup mieux.
# Séparons les données en deux catégories : hommes et femmes.
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips_dataset, jitter = True, hue = 'sex')

# Homme et femme combinés pour chaque jour.
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips_dataset, jitter = True, hue = 'sex', dodge = True)

# Hommes et femmes séparés pour chaque jour
# split parameter has been renamed to 'dodge'
```

Il est maintenant temps d'essayer un swarmplot.

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips_dataset)

#Similaire à Strip Plot.
# Peut se heurter à des problèmes de calcul et de visualisation avec de très grands ensembles de données
```

Nous pouvons combiner un Violin Plot et un Swarm Plot dans le même graphique. Essayons ici en mettant 2 lignes de code dans la même cellule.

```python
sns.violinplot(x= 'day', y= 'total_bill', data= tips_dataset)
sns.swarmplot(x= 'day', y= 'total_bill', data= tips_dataset, color= 'black')

# Ici, nous avons changé la couleur de la parcelle de Swarm Plot pour une meilleure visibilité.
# Les graphiques Violon et Swarm donnent une bonne vue de la distribution et de la densité des données.
```

Now let’s try the Factor Plot

```python
sns.factorplot(x = 'day', y = 'total_bill', data = tips_dataset)

# C'est le visuel par défaut si vous ne spécifiez pas le "type" de visuel.
```

```python
sns.factorplot(x = 'day', y = 'total_bill', data = tips_dataset, kind = 'bar')
```

Enfin, nous allons explorer Facet Gid()

```python
sns.FacetGrid(data= tips_dataset, row= 'time', col= 'smoker')

# Ceci initialise la grille. Maintenant, ajoutons quelques visuels d'histogramme.
```

```python
(sns.FacetGrid(data= tips_dataset, row= 'time', col= 'smoker')).map(plt.hist,'total_bill')

# Nous explorons ici un histogramme des fumeurs par rapport aux non-fumeurs pendant le déjeuner et le dîner
```

Faisons un style de Scatter Plot pour FacetGrid()

```python
graph= sns.FacetGrid(data= tips_dataset, row= 'time', col= 'smoker')
graph.map(plt.scatter, 'total_bill', 'tip', edgecolor= 'w')

# Ici nous explorons les fumeurs vs les non-fumeurs qui paient des pourboires pendant le déjeuner vs le dîner.
```

```python
graph= sns.FacetGrid(data= tips_dataset, col= 'time', hue= 'smoker')
(graph.map(plt.scatter, 'total_bill', 'tip', edgecolor= 'w')).add_legend()

# Ici, nous attribuons une des variables (fumeur) à la couleur des données de la parcelle.
```
