```python
from requests import get
```

```python
import pandas as pd
```

```python
chemin = "https://www.lequipe.fr/Football/FootballFicheClub26.html"
```

```python
page = get(chemin)
if page.status_code == 200:
    print("Tout s'est bien passé")
else:
    print("Problème de récupération")
```

```python
texte = page.text
```

```python
import re
```

```python
motif = re.compile("[Cc]alendrier [Pp]aris-[Ss]g")
```

```python
resultats = motif.findall(texte)
print(type(resultats))
print(len(resultats))
```

```python
for resultat in resultats:
    print(resultat)
```

```python
motif = re.compile("[Cc]alendrier [Pp]aris-[Ss]g(.{1,50})")
resultats = motif.findall(texte)
print(type(resultats))
print(len(resultats))
```

```python
for resultat in resultats:
    print(resultat)
```

```python
motif = re.compile("[Cc]alendrier [Pp]aris-[Ss]g((.|\\n){1,50})")
resultats = motif.findall(texte)
print(type(resultats))
print(len(resultats))
```

```python
for resultat in resultats:
    print(resultat)
```

```python
import bs4
```

```python
arbre = bs4.BeautifulSoup(texte, "lxml")
```

```python
noeud = arbre.h1.text.replace("\n", "")
```

```python
print(noeud)
```

```python
div_fille = arbre.find(id="container").find(class_="Palmares calendrierclub")
str(div_fille)[0:1000]
```

```python
all_ads = div_fille.find("table").findAll("tr")
print(len(all_ads))
```

```python
list_all_ads = []
for ads in all_ads:
    date = ads.find_next(class_="date")
    competition = ads.find_next(class_="competition")
    equipe1 = ads.find_next(class_="equipe equipe1")
    score = ads.find_next(class_="score")
    equipe2 = ads.find_next(class_="equipe equipe2")
    list_all_ads = []
for ads in all_ads:
    date = ads.find_next(class_="date")
    competition = ads.find_next(class_="competition")
    equipe1 = ads.find_next(class_="equipe equipe1")
    score = ads.find_next(class_="score", )
    equipe2 = ads.find_next(class_="equipe equipe2")
    list_all_ads.append({"mois":pd.to_datetime(date.text).month,
                         "date":pd.to_datetime(date.text),
                         "equipe1":equipe1.text.replace(" ","") ,
                         "competition":competition.text[0:2],
                         "score":score.text,
                         "equipe2":equipe2.text.replace(" ",""),
                         "equipe1_goal":pd.to_numeric(score.text[0]),
                         "equipe2_goal":pd.to_numeric(score.text[2]),
                        })
```

```python
df_match = pd.DataFrame(list_all_ads)
df_match
```

```python
df_match.shape
```

```python
df_match = df_match.drop([0,0],axis=0)
df_match
```

```python
df_match.loc[(df_match.equipe1 == "Paris-SG") & (df_match.equipe1_goal > df_match.equipe2_goal), 'Bilan'] = 'V'
df_match.loc[(df_match.equipe1 == "Paris-SG") & (df_match.equipe1_goal < df_match.equipe2_goal), 'Bilan'] = 'D'
df_match.loc[(df_match.equipe1 == "Paris-SG") & (df_match.equipe1_goal == df_match.equipe2_goal), 'Bilan'] = 'N'

df_match.loc[(df_match.equipe2 == "Paris-SG") & (df_match.equipe1_goal > df_match.equipe2_goal), 'Bilan'] = 'D'
df_match.loc[(df_match.equipe2 == "Paris-SG") & (df_match.equipe1_goal < df_match.equipe2_goal), 'Bilan'] = 'V'
df_match.loc[(df_match.equipe2 == "Paris-SG") & (df_match.equipe1_goal == df_match.equipe2_goal), 'Bilan'] = 'N'
df_match
```

```python
df_match.info()
```
